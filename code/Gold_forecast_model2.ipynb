```python
#!/usr/bin/env python
#coding=utf-8
temp = 0
import pandas as pd
#从文件里得到数据
df = pd.read_csv('./data/LBMA-GOLD_test.csv',encoding='gbk')
#maIntervalList = [30,3]
df.columns = ['Date','exchange rate']
#虽然在后文里只用到了5日均线，但这里演示设置3种均线
#for maInterval in maIntervalList:
#    df['MA_' + str(maInterval)] = df['exchange rate'].rolling(window=maInterval).mean()
cnt=0    
print(gold)
while cnt<=len(df)-1:
     try:
        if df.iloc[cnt+1]['exchange rate']<df.iloc[cnt]['exchange rate'] \
        and (df.iloc[cnt]['exchange rate']-df.iloc[cnt+1]['exchange rate'])>df.iloc[cnt]['exchange rate']*0.01:

                    if gold>0 :
                         print("Sell Point on:" + df.iloc[cnt]['Date'])
                         temp = gold*0.2
                         #temp*df.iloc[cnt]['exchange rate']*0.01
                         gold = gold-temp
                         profit =profit +  temp*df.iloc[cnt]['exchange rate']
                         profit = profit - temp*df.iloc[cnt]['exchange rate']*0.01
                         #print(temp)
                         print(gold)
                         print(profit)
                         ls1.append(df.iloc[cnt]['Date'])
                         ls2.append(1)

                    
       #规则1，收盘价连续三天上扬
    
     #   elif (df.iloc[cnt]['exchange rate']<df.iloc[cnt+1]['exchange rate'] and df.iloc[cnt+1]['exchange rate']<df.iloc[cnt+2]['exchange rate']) \
     #       or  (df.iloc[cnt]['MA_3']<df.iloc[cnt+1]['MA_3'] and df.iloc[cnt+1]['MA_3']<df.iloc[cnt+2]['MA_3']) \
      #      or (df.iloc[cnt+1]['MA_3']>df.iloc[cnt]['exchange rate'] and df.iloc[cnt+2]['MA_3']<df.iloc[cnt+1]['exchange rate']):

        elif df.iloc[cnt+1]['exchange rate']>df.iloc[cnt]['exchange rate'] \
        and (df.iloc[cnt+1]['exchange rate']-df.iloc[cnt]['exchange rate'])>df.iloc[cnt]['exchange rate']*0.01:
    
                    
                    if profit>0:
                        print("Buy Point on:" + df.iloc[cnt]['Date'])
                        temp = profit*0.2
                        profit = profit - temp
                        gold = gold + temp*(1/df.iloc[cnt]['exchange rate'])
                        profit = profit - temp*0.01
                         #print(temp)
                        print(gold)
                        print(profit)
                        ls2.append(2)
                        ls1.append(df.iloc[cnt]['Date'])

     except: 
         pass
     cnt=cnt+1
     #ls1.append(profit)
     #ls2.append(gold)
     ls3.append(profit+gold*df.iloc[cnt-1]['exchange rate'])
     
    
total_profit=profit+gold*df.iloc[cnt-1]['exchange rate']
print(total_profit)
```
